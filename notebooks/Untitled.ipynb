```python
import sys
import base64
import os
sys.path.append('..')
import blacknet.settings as s
```

```python
fp = os.path.join(s.GUITARPRO_DIR, 'posthardcore', 'Silverstein - Smashed Into Pieces.gp4')
```

```python
def read_bytes_to_int(f, n_bytes=1):
    b = f.read(n_bytes)
    return int.from_bytes(b, byteorder='little')

def read_character_string(f):
    n_chars = read_bytes_to_int(f)
    return f.read(n_chars).decode('utf-8')

def read_delphistring(f):
    maxl = read_bytes_to_int(f, 4)
    l = read_bytes_to_int(f, 1)
    
    if maxl != l + 1 and maxl > 255:
        l = maxl - 1
        
    # print(maxl)
    # print(l)
        
    if l > 0:
        s = f.read(l).decode('utf-8')
    else:
        s = ''
    return s
```

```python
with open(fp, 'rb') as f:
    
    ## HEADERS
    # read version
    string_size = read_character_string(f)
    print('VERSION:\t{}'.format(string_size))
    
    # read empty bytes till 30th byte
    f.read(30 - f.tell())
    read_bytes_to_int(f)
    
    print(f.tell())
    
    # read info
    title = read_delphistring(f)
    print('TITLE:\t{}'.format(title))
    subtitle = read_delphistring(f)
    print('SUBTITLE:\t{}'.format(subtitle))
    artist = read_delphistring(f)
    print('ARTIST:\t{}'.format(artist))
    album = read_delphistring(f)
    print('ALBUM:\t{}'.format(album))
    composer = read_delphistring(f)
    print('COMPOSER:\t{}'.format(composer))
    copyright = read_delphistring(f)
    print('COPYRIGHT:\t{}'.format(copyright))
    transcriber = read_delphistring(f)
    print('TRANSCRIBER:\t{}'.format(transcriber))
    instructions = read_delphistring(f)
    print('INSTRUCTIONS:\t{}'.format(instructions))
    n_comments = read_bytes_to_int(f, 4)
    comments = [read_delphistring(f) for _ in range(n_comments)]
    print('COMMENTS:\t{}'.format(comments))
    
    # Triplet feel
    triplet_feel = bool(read_bytes_to_int(f, 1))
    print('TRIPLET FEEL:\t{}'.format(triplet_feel))
    
    # Lyrics
    lyrics_track = read_bytes_to_int(f, 4)
    print('LYRICs TRACK:\t{}'.format(lyrics_track))
    lyrics_from_beat = read_bytes_to_int(f, 4)
    print('LYRICS FROM BEAT:\t{}'.format(lyrics_from_beat))
    
    print(f.tell())
```
